```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plane_wave_1d(omegas,x):
    psi_x = np.zeros(len(x), dtype=complex)
    for omega in omegas:
        for idx,xi in enumerate(x):
            psi_x[idx] += np.exp(1j*omega*xi)
    return psi_x/(2*np.pi) 
```

```python
x_min = -5
x_max = 5
max_freq = 10
nsteps = 100
nfreqs = 20
x = np.array(range(x_min*nsteps, x_max*nsteps))/nsteps
y = plane_wave_1d(np.array(range(0,int(max_freq*nfreqs)))/nfreqs,x)
```

```python
plt.plot(x,np.real(y))
plt.plot(x,np.imag(y))
```

```python
%matplotlib qt
# %matplotlib inline 

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(x, np.imag(y), np.real(y));
# ax.scatter3D(x, np.imag(y), np.real(y), c=x, cmap='Greens');
ax.set_xlabel('x')
ax.set_ylabel('Real(y)')
ax.set_zlabel('Imag(y)')

plt.show()
```
